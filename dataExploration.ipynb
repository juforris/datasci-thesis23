# Data exploration

```python
# Data wrangling and plotting
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Load data

```python
fragment_file = "/mnt/project/Transpose/hidra/results/salmon-hidra/batch1/fragment_filtered_fasta/filtered_fragments.fasta"
deseq_file = "/mnt/project/Transpose/hidra/results/salmon-hidra/batch1/DESeq_TSV/DESeq_res.tsv"

fragments = pd.read_csv(fragment_file, header=None, names=["fragment","sequence"], sep="\t")
deseq = pd.read_csv(deseq_file, sep="\t")
```

### Fragment lengths

```python
fragments["length"] = fragments.sequence.str.len()
```

```python
fragments.length.hist(grid=False, bins=70, color='#66856a')
plt.title("Fragment length distribution")
plt.xlabel("Bases")
plt.savefig("fragment_lengths.pdf")
plt.show()
```

```python
fragments.length.describe()
```

```python
np.median(fragments.length)
```

### Base mean distribution

```python
sns.set(font_scale=2)
sns.set_style('whitegrid')
plt.figure(figsize=(20,2))
sns.boxplot(x=deseq.baseMean, color=sns.color_palette("Pastel2")[4], showfliers = False)
plt.title("Base mean distribution")
plt.xticks(np.arange(0,95,5))
plt.xlabel("Base mean")
plt.savefig("basemeanDist.pdf", bbox_inches="tight")
```

### Log 2 Fold Change distribution

```python
sns.set(font_scale=2)
sns.set_style('whitegrid')
plt.figure(figsize=(20,2))
sns.boxplot(x=deseq.log2FoldChange, color=sns.color_palette("Pastel2")[0], showfliers=False)
plt.title("Log2 Fold Change distribution")
plt.xticks(np.arange(-1.25, 1.5, 0.25))
plt.xlabel("log2 Fold Change")
plt.savefig("log2fcDist.pdf", bbox_inches="tight")
```
